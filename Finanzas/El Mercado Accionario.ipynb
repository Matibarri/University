## Bolsa de Comercio

- En Chile, se transa en la Bolsa de Santiago. www.bolsadesantiago.com

- En 1893 se fundó la Bolsa de Comercio de Santiago.

- En 1898 fue creada la Bolsa de Valores (de Corredores) de Valparaíso. Inicialmente las transacciones eran mayores que en Santiago por la actividad comercio del puerto. Se cerró en 2018

- En 1989 se creó la Bolsa Electrónica de Chile.

### Empresas que transan en bolsa

- **Sociedades abiertas**: con $500$ o más accionistas, o que al menos el $10\%$ de su capital suscrito pertenece a un mínimo de 100 accionistas.

- **Sociedades cerradas**: que voluntariamente deciden inscribir sus acciones en el Registro de Valores, y por lo tanto pueden ser transadas en la bolsa.

### Administradores de las sociedad

- **Junta de Accionistas**: órgano conformado por todos los accionistas y que toman las decisiones clave del funcionamiento de la sociedad.

- **El Directorio**: ejerce la administración de la S.A.

- **El Gerente General**: administrador de la empresa designado/elegido por el directorio.

La decisión, por ejemplo, de cuántos dividendos repartir lo propone el Gerente General, pero se aprueba en el Directorio.
Si un inversionista tiene propiedad de una acción, tiene dos tipos de ganancias:

- **Ganancia económica-financiera de ser accionista**: Dividendos más ganancia de capital.

- **Ganancia administrativa de ser accionista**: Control de la empresa.

Matemáticamente, la ganancia económica se define como:

Ganancia económica $= \sum_{n=1}^N \frac{BN_n}{(1+r)^n} + [\frac{P_N}{(1+r)^n} - P_0]$

$ \sum_{n=1}^N \frac{BN_n}{(1+r)^n} =$ suma actualizada de dividendos recibidos.

$[\frac{P_N}{(1+r)^n} - P_0] =$ ganancia de capital por sobreprecio de venta. (diferencia entre que se vende la acción y que se compra)

Desde la perspectiva de la empresa, como se supone que se trabaja con accionistas de largo plazo, entonces $P_N$ tiende a $0$. Sin embargo, desde la perspectiva del accionista, el decide como decisión endógena, cuándo vende su acción.

### Corredoras de Bolsa

Para comprar o vender acciones en la Bolsa, se debe hacer por medio de las corredoras de Bolsa. Cuando se transan acciones, por este servicio de intermediación financiera, tanto la empresa que emite acciones como el inversionista compra acciones, deben pagar una comisión.
Comisiones en compra-venta de acciones:

- Comisión de la corredora: $0.35 \% - 1\%$

    - Derecho bolsa: $0.5\%$ (precio puesto por la Bolsa es fijo)
    - IVA (sobre la suma anterior)

- Total Comisión: $0.5\% - 1.5 \%$ (Es lo que denominamos $c$ en el modelo anterior).

**Ejemplos de corredoras**: LarrainVial, BanChile corredora, EuroAmerica, BancoEstado, Ureta y Bianchi, CelfinCapital, Munita, Cruzat y Claro.

### IPSA

Con respecto a la caracterizaicón del mercado vurstátil en Chile, una forma de saber si la bolsa "va bien o va mal" es ver un índice que se llama **Índice de Precios Selectivo de Acciones (IPSA)**. Básicamente es un resumen que indica la rentabilidad de las $30$ acciones más transadas en Chile.

- Creado en 1977.
- Indicador de rentabilidad, que mida las variaciones de precios de las $30$ acciones de las sociedades más transadas (actualizado al $2020$, antes eran las $40$ acciones más transadas).
- La cartera del IPSA se reevalúa el última día hábil de cada año.

### IGPA

El **Índice General de Precios de Acciones (IGPA)**, creado en $1958$ es un indicador de carácter patrimonial, que mide las variaciones de precios de la mayoría de las acciones inscritas en Bolsa, es general, incluye a todas las acciones que están inscritas en la Bolsa de Santiago. Dicha medición se efectúa a través del **Patrimonio Bursátil** (N° de acciones X valor actual de ellas).
Dicho índice es representativa del mercado, ya que incluye a todas las acciones. Sin embargo, como hay acciones que no son frecuentemente transadas o se transan muy poco, el IGPA no fluctúa tanto.

**Observación**: Dependerá del objetivo del análisis para considerar el **IGPA** o el **IPSA**.

**Rubros del IGPA**

1. Bancarias y financieras.
2. Agropecuarias y forestales.
3. Mineras.
4. Industriales.
5. Servicios básicos.

Detallando el punto 4. Los subrubros industriales son:

 1. Alimentos y bebidas.
 2. Construcción.
 3. Productos diversos.
 4. Metalmecánicas.
 5. Pesqueras.
 6. Productos químicos.
 7. Textiles y vestuarios.

Detallando el punto 5. Los subrubros de servicios básicos som:

  1. Comerciales y distribuidoras.
  2. Inversiones e inmobiliarias.
  3. Marítimas y navieras.
  4. Servicios públicos.
  5. Seguros.
