## Modelo CAPM para medir costo de capital

### Introducción

El costo de capital implícito está determinado por el rendimiento que se puede obtener por inversiones efectuadas fuera de la empresa.
Estudiaremos un modelo que captura el comportamiento del mercado y nos permite obtener una tasa de descuento para los propietarios.

El modelo CAPM se deduce a partir del modelo muy conocido llamado el modelo de Markowitz y del Modelo de Sharpe.
Referencias:

- Markowitz, H.(1959). Portfolio selection: efficient diversification of investments (Vol.16). Wiley

- Sharpe, W.F. (1963). A simplified model for portfolio anaylisis. Management Science, 9(2), 277-293.

- Sharpe, W.F. (1964). Capital asset prices: A theory of market equilibrium under conditions of risk. The Journal of Finance, 19(3), 425-442.

### Modelo

El modelo CAPM es representado por la siguiente ecuación:

$R_i=R_L + \beta(R_m - R_L)$

donde:
- $R_i =$ rentabilidad del activo financiero $i$ o rentabilidad de las acciones de una empresa $i$.
- $R_L =$ tasa libre de riesgo.
- $R_m =$ rentabilidad del portafolio de mercado.
- $\beta =$ premio por riesgo. (coeficiente de volatibilidad)

La idea es estimar esta ecuación para obtener la rentabilidad exigida por una determinada empresa $i$.
La especificación econométrica del CAPM es representado por la siguiente ecuación:

$R_i-R_L= \alpha + \beta(R_m - R_L) + \epsilon_i$

donde: 
- $R_i =$ rentabilidad del activo financiero $i$ o rentabilidad de las acciones de una empresa $i$.
- $R_L =$ tasa libre de riesgo. (proxy: activos poca varianza)
- $R_m =$ rentabilidad del portafolio de mercado. (proxy: índices de mercado)
- $\beta =$ premio por riesgo. (coeficiente de volatibilidad)
- $\epsilon_i =$ error estadístico del activo $i$ (características no observadas)

Si se considera $R_i - R_L = Y_i$ como la variable dependiente y $R_m - R_L$ la variable explicativa del modelo, entonces:

$Y_i = \alpha + \beta X_i + \epsilon_i$

Del cual se estiman $\alpha,\beta$, donde nos interesa $\hat{\beta}$, lo que nos va a permitir predecir cuál es la rentabilidad esperada de cierto activo $i$ ($R_i$) y esto nos va a indicar cuál es la exigencia de capital.
Además, se utiliza como supuesto que $\epsilon_i$ se distribuye Normal.

El modelo CAPM indica que la rentabilidad de un activo financiero $i$, en equilibrio, se obtiene de este modelo teniendo como datos conocidos los siguientes:

- Tasa libre de riesgo. ($R_L$: normalmente se usa una variable proxy/aprox. con una varianza cercana a 0)
- Beta de la empresa. ($\hat{\beta}$)
- Rentabilidad del portafolio de mercado. ($R_m$) 

Por ser una rentabilidad de equilibrio, entonces los accionistas de una empresa exigirán como costo de capital esta misma tasa de rentabilidad, es decir:

$\hat{R_i} = R_L + \hat{\beta}(R_m - R_L) = k_p$

con $K_p$: costo implícito del capital.

Esto implica que el propietario exigirá $K_p$ considerando características como la rentabilidad que puede tener en otras alternativas financieras y la rentabilidad de los activos libres de riesgo.

### Rentabilidad del activo financiero $i$:

$R_{i,j}=\frac{P_{i,j} - P_{i,j -1}}{P_{i,j -1}}$ 

donde:

- $R_{i,j} =$ rentabilidad del activo financiero $i$ en el periodo $j$.
- $P_{i,j} =$ precio de mercado del activo financiero $i$ en el periodo $j$.
- $P_{i,j} =$ precio del activo financiero $i$ en el periodo anterior a $j$ o $j-1$.

Diferencia entre los precios en dos períodos de tiempo. Como la bolsa normalmente es fluctuante, es apropiado trabajar con datos diarios si se trabaja con una acción que se transa mucho en la bolsa (variación de un día a otro). Ahora, si la acción se transa menos y varía menos, se puede trabajar con otros períodos de tiempo como mensuales, trimestrales, anuales, etc.

Nótese que en $R_{i,j}$ no se incorporan los dividendos que entrega la acción, ya que esto ya está incluída en sí misma.

### Rentabilidad de un activo libre de riesgo ($R_L$)

- Rentabilidad que se obtiene en una inversión que no tiene riesgo.
- Plantea la idea de costo alternativo, en una inversión relativamente segura.
- En la vida real, se observan activos con varianza muy baja: Bonos de Tesorería de la República, Bonos del Banco Central.
- Es muy difícil encontrar un activo que tenga rentabilidad con varianza 0, es decir, que no tenga riesgo.
- Normalmente, en Chile se usa la rentabilidad de los Bonos del Banco Central como un buen proxy del $R_L$.

### Portafolio de mercado

- El portafolio de mercado debería tener varios activos, no sólo financieros.

- En la vida real, tal portafolio es difícil de encontrar.

- Como proxy se usan indicadores como los índices de precios accionarios.

- Estos indicadores muestran la evolución del mercado. Para Chile, puede ser el Índice General de Precios de Acciones (IGPA) o el Índice de Precios Selectivos de Acciones (IPSA).

El portafolio de mercado es una canasta de inversiones que incluye todas las inversiones que tiene una determinada persona.

### Rentabilidad de mercado ($R_m$)

Si se usa el IPSA, la rentabilidad de mercado es:

$R_m = \frac{{IPSA}_t - {IPSA}_{t-1}}{{IPSA}_{t-1}}$

**Recordar**: El IPSA calcula la evolución del precio de las 30 acciones más transadas de la Bola Chilena. También puede ser diaria, semanal, mensual, etc.

**Importante**: Algunos estudios toman las acciones más transadas del IGPA. Así, se podría construir un portafolio de mercado aproximado distinto al portafolio de mercado aproximado que construye el IPSA. Sin embargo, los valores suelen ser muy similares, ya que muchas de las acciones más transadas IGPA coinciden con las del IPSA.

### Especificación empírica modelo CAPM:

Recordando el modelo, se tiene que:

$R_{i,t} - R_{L,t} = \alpha + \beta_t(R_{m,t} - R_{L,t}) + \epsilon_{i,t}$

donde $t$ representa un momento del tiempo (día, meses, años, etc.) Además, $\beta$ se puede estimar econométricamente, siendo el método más simple el de mínimos cuadrados ordinarios (MCO).

Por ejemplo, Falabella. Tomamos una serie de precios de la empresa Falabella, tomamos una serie de rentabilidades de riesgo y una serie de rentabilidades del IPSA, entonces podemos estimar cuál es el premio por riesgo que nos entregan las inversiones de Falabella a través de sus activos financieros. Según eso, podemos calcular cuál es la rentabilidad que un propietario debería exigirle a esa empresa, en este caso, Falabella.

**Observación**: Esto funciona para empresas que transan en la Bolsa. La rentabilidad de mercado captura todas las alternativas de inversión que tiene una inversionista y además la tasa libre de riesgo aproximada empíricamente hace de cuota inferior a la rentabilidad que le va a exigir a la empresa.

**Nota**: No funciona para empresas que no transen en la Bolsa, ya que no vamos a poder tener los datos de rentabilidad de ese activo $i$ ($R_i$), esto implica que no se podrá construir la variable dependiente. Por lo tanto, lo que se hace es utilizar $\beta$ de empresas similares que sean del mismo rubro, mismo tamaño, que operen en los mismos mercados, etc.

Tener muy en cuenta los supuestos que se estén usando al momento de realizar análisis.

### Determinación del costo implícito

Estimando el $\hat{\beta}$, se reemplaza en la ecuación del modelo, obteniendo así el costo implícito.

$\hat{k_p} = R_L + \hat{\beta}(R_m - R_L)$

Por lo tanto, teniendo $\hat{\beta}$, $R_L$ y $R_i$ es posible calcular el $k_p$ para los propietarios de una empresa.

### Recordar: Valoración de empresas

Valor presente de los flujos de caja futuros esperados:

$VP = \frac{{Flujo de caja neto}_1}{(1+i)^1} + \frac{{Flujo de caja neto}_2}{(1+i)^2} + ... + \frac{{Flujo de caja neto}_t}{(1+i)^t}$

- **Flujo de cada neto**: neto de impuesto e inversiones requeridas para mantener o aumentar la capacidad de generar flujos de caja futuros.

- Tasa de descuento relevante se obtiene usando el Costo Capital Promedio Ponderado (WACC) o Modelo de Valoración del Precio de los Activos Financieros (CAPM).

Luego de haber determinado el costo implícito, se tiene que:

$\hat{k_p} = i$ = Tasa de descuento

**Observación**: Es apropiado usar el modelo CAPM cuando la empresa transa en Bolsa o cuando la empresa que no transan en Bolsa, pero que tienen una contraparte de una empresa que sí lo hace. Así, en caso contrario, será más apropiado será usar el WACC.

**Nota**:Hay ocasiones que lo que se hace es calcular un costo implícito usando el modelo CAPM muy aproximado, considerando el beta de todas las empresas del sector que sí transan en Bolsa (beta promedio) y también calcular del WACC. Luego, se observa cómo van cambiando los VP con esas tasas de descuento. A veces cambian poco, otras cambia mucho. El promedio de los betas podría ser $\bar \beta \pm \sigma(\bar \beta)$, dando así un intervalo para el $\hat{k_p}$.

### Beta

Al ser beta la representación del premio por riesgo, van a ver empresas que pagan más por el riesgo, la cual dependerá de la rentabilidad de cada empresa. Sin embargo, también depende de qué tan volátil sea.

- Si $\beta > 1$ la empresa es volátil en sus acciones, es decir, por unidad de riesgo que se tome, la empresa entregará "más" que $1$.

- Si $\beta < 1$ la empresa no es volátil en sus acciones, es decir, por unidad de riesgo que se tome, la empresa entregará "menos" que $1$.

Se pueden encontrar información de betas en páginas financieras, por ejemplo, Emol Inversiones. Con este dato, junto con la tasa libre de riesgo y la rentabilidad de mercado, el accionista puede determinar el costo implícito del financiamiento.

### Cambios en la estructura deuda-patrimonio

Cuando se calcula el beta a partir del modelo indicado y tomando los datos de cada empresa, se recogen las características de dicha empresa. Sin embargo, puede haber cambios en la estructura de financiamiento, por ejemplo, podría haber nuevo endeudamiento o nuevos aportes de capital que el beta histórico no muestra. Por ello, se debe plantear cuál será el nuevo coeficiente beta que recoja los datos de financiamiento que se pretende desarrollar.

### Beta de empresas con endeudamiento

Es posible plantear una relación entre un beta de una empresa con deuda y sin deuda.

$\beta_D = (1 + (1-t)\frac{D}{P})\beta_A$

donde:

- $\beta_D =$ beta empresa con deuda.
- $\beta_A =$ beta empresa sin deuda.
- $\frac{D}{P} =$ relación deuda-patrimonio, ambios (deuda y patrimonio) a precio de mercado.
- $t =$ tasa de impuestos a las utilidades de la empresa.

Observar que el beta con deuda es mayor que el beta sin deuda o autofinanciada. La diferencia entre ambos se debe al riesgo financiero que implica una deuda, es decir, el premio por riesgo es mayor cuando la empresa está endeudada (especio de comprensación por estar dispuestos a invertir en empresas que son más riesgosas)

**Observación**: El beta no indica si una empresa es atractiva o no para invertir, sino que lo que indica es cuánto premia por unidad de riesgo. Para analizar de forma completa cuál es ese premio por unidad de riesgo, es necesario analizar la estructura de dendeudamiento de la empresa.

## Estimación Econométrica del CAPM

Al momento de realizar una estimación, la Rentabilidad del mercado se puede considerar el IPSA para realizar los cálculos. La tasa libre de riesgo se puede considerar como la tasa de interés o la rentabilidad de un bono del Banco Central, por ejemplo, el bono a 2 años sería una tasa de interés diaria. Así, también habría que trabajar con el IPSA diario (y todo en general a diario).

El modelo de estimación es la regresión lineal por MCO:

$Y_i = \alpha + \beta X_i + \epsilon_i$

Así, los estimadores del MCO de $\alpha$ y $\beta$ son:

$\hat{\alpha}_{MCO}= \bar Y - \beta \bar X$

$\hat{\beta}_{MCO}= \frac{\sum_{i=1}^{N} (x_i -\bar x)(y_i - \bar y)}{\sum_{i=1}^{N} (x_i -\bar x)^2} $

**Observación**: Al presentar los resultados se debe editar la tabla, es decir, no copiar y pegar el output del programa. Un ejemplo de tabla se presenta en la diapositiva 21 del documento "Estimación del CAPM".



**Costo implícito estimado**

Supongamos que se estimó el beta, resultando $\beta = 0.6289$. Luego, la predicción del costo implícito será:

$\hat{k_p} = R_{L,t} + 0.6289(R_{m,t} - R_{L,t})$

Con esto, se puede calcular $\hat{k_p}$ para cualquier tasa libre de riesgo y rentabilidad de mercado.

### Modelo CAPM para distintos horizontes de tiempo.

Fernández, Viviana (2005). El modelo CAPM para distintos horizontes de tiempo, revista Ingeniería de Sistemas, Vol.19.

- Análisis para 24 empresas que transan en la Bolsa de Comercio de Santiago con más de $87\%$ de presencia bursátil entre $1997 - 2002$.

- Para un horizonte de $16 - 32$ días, beta promedio de $0.75$.

- Para un horizonte de $64 - 128$ días, beta promedio de $0.97$.

Por lo tanto, hay que ser claros en el horizonte de tiempo que se está considerando en la estimación.
