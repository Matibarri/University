# La globalización, mercados en moneda extranjera y riesgo cambiario

## Globalización

Es un proceso de integración de países de todo el mundo que incluye a lo menos cuatro componentes:

- Aumento de comercio internacional.
- Aumento de flujos financieros internacionales.
- Mayor internalización de los procesos productivos.
- Avance hacia la armonización de las instituciones económicas internacionales.

**Finanzas Internacionales**: Área de la finanzas que estudia el comercio internacional y flujos financieros internacionales.

## Riesgos del comercio internacional

- Asociados de la incertidumbre con los tipos de cambios: Variaciones no esperadas en los tipos de cambio puede afectar de distinta manera a los ingresos (efecto en empresas exportadoras), costos (efecto en empresas importadoras) y utilidades de la empresa.

- Otro riesgo típico es el riesgo país: No recibir pago de exportaciones por eventos políticos, económicos, social, etc. Como también la incertidumbre respecto a la imposición, cambios en tarifas, cuotas de importación, otras barreras no tarifarias, etc.

Las finanzas de empresas se preocupa netamente de la incertidumbre con los tipos de cambios.

## Mercados en moneda extranjera

- Tipo de cambio para billetes bancarios: por ejemplo, billetes de la Reserva Federal (USA) y billetes del Bank of England.

- Tipo de cambio para los cheques que estipulan montos en dólares y en otra moneda: dependen de si éstos son emitidos por bancos o corporaciones, los montos y las fechas.

- Tipo de cambio de moneda extranjera al contado: el que típicamente conocemos como "tipo de cambio".

## Mercado de moneda extranjera al contado:

- El tipo de cambio al contado (precio) se determina a través de la oferta y la demanda para las divisas que se negocian en un mercado global intercambiario de divisas. (Divisa = Moneda que se acepta en términos internacionales como el dólar, euro o el yuan)

- El mercado intercambiario de divisas es el mercado financiero más grande del mundo: Rotación diaria de casi un billón de dólares.

## Características del mercado cambiario al contado:

- La eficiencia del mercado cambiario al contado se observa en el bajo diferencial cambiario que existe entre el precio de venta y de compra ($<$ al $1\%$).
- Opera en dos niveles:

    1. Nivel directo interbancario: bancos negocian entre sí.
    2. Nivel indirecto por vía de corredores: bancos colocan (divisas) con los corredores. Corredores sólo pueden hacer trato con terceros.
    
- Los bancos y corredores se conectan entre sí a través de una red de comunicaciones por satélite denominada **Societe for Worldwide International Financial Telecommunications (SWIFT)**. Basado en Bruselas, Bélgica e interconecta a los bancos y a los corredores de casi todo centro financiero. Esto ayuda a los altos niveles de eficiencia que tiene el mercado cambiario.

## Exposición y Riesgo cambiario

### Definiciones

- **Riesgo cambiario**: Variabilidad de los valores de los activos, de los pasivos o de los ingresos en operación en moneda nacional que resulta de las variaciones no anticipadas en los tipos de cambio.

- **Exposición cambiaria**: Aquella parte de los activos, pasivos o ingresos que se encuentran sujetos a riesgo.

- **Nivel de exposición cambiaria**: Sensibilidad de las variaciones en el valor real de la moneda nacional de los activos, pasivos y de los ingresos, en operación, en función de las variaciones no anticipadas en los tipos de cambio. 

### Estimación de la exposición cambiaria

Relación entre tipo de cambio y valores de la empresa:

Se puede definir una relación lineal entre las variaciones inesperadas en los tipos de cambio ($\Delta$TC) y el cambio en el valor real de los activos, pasivos o ingresos operacionales individuales y particulares o como el cambio en el valor real de un conjunto de activos, pasivos o ingresos de operacionado (se denota este cambio como $\Delta$V).

Se modelará los cambios en los valores de los activos, pasivos o de los ingresos de operación ($\Delta$V) como la dependencia de las variaciones en el tipo de cambio($\Delta$TC) y de otros factores ajenos al tipo de cambio denotados por $\alpha$ que representan la constante del modelo estimado.

Se puede estimar la exposición cambiaria por MCO de la siguiente manera:

$\Delta V = \beta \Delta TC + \alpha + \epsilon$

donde:

- $\Delta V =$ cambios en los valores de los activos, pasivos o ingresos de operación.
- $\Delta TC =$ variaciones no anticipadas en el tipo de cambio.
- $\alpha =$ otros factores ajenos al tipo de cambio.
- $\epsilon =$ error estocástico (características no observadas)

Parámetro a estimar:

$\beta =$ exposición cambiaria. Describe la sensibilidad entre variaciones no anticipadas en los tipos de cambio y $\Delta$V.

La relación anterior se podría extender a múltiples tipos de cambio. Así, se podría definir lo siguiente a modo ejemplo:

$\Delta V = \beta_1 \Delta TC(pesos/dólar) + \beta_2 \Delta TC(pesos/libras) + \beta_3 \Delta TC(pesos/yen) + \beta_4 \Delta TC(pesos/franco suizo) \alpha + \epsilon$

donde:

- $\beta_1 =$ sensibilidad de $\Delta TC$ a variaciones no anticipadas en el valor del peso expresado en dólares.
- $\beta_2 =$ sensibilidad de $\Delta TC$ a variaciones no anticipadas en el valor del peso expresado en libras esterlinas.
- $\beta_3 =$ sensibilidad de $\Delta TC$ a variaciones no anticipadas en el valor del peso expresado en yenes.
- $\beta_4 =$ sensibilidad de $\Delta TC$ a variaciones no anticipadas en el valor del peso expresado en francos suizos.
- $\alpha =$ otros factores no considerados que afectan a $\Delta TC$.
- $\epsilon =$ error estocástico (características no observadas)

### Riesgo cambiario

**Medición riesgo cambiario**: Se mide por medio de la varianza del valor de un activo, de un pasivo o de un ingreso operacional expresado en modena nacional y que es atribuible a las variaciones no anticipadas en el tipo de cambio.

Se centra sobre la impredecibilidad de los valores **ocasionados** por la incertidumbre en los tipos de cambio y no por la incertidumbre en los tipos de cambios en sí mismo.

Matemáticamente, el riesgo cambiario se expresa como:

var($\hat{V}) =$ var($\beta \Delta$TC) $= \beta^2$var($\Delta$TC)

donde:

- var $=$ varianza.
- $\hat{V} =$ valor de un activo, pasivo o ingreso operacional sujeto a tipo de cambios.

Usualmente, el Riesgo de tipo de cambio está en descritos y cuantificados en los EEFF (balances) de las empresas.

**Mitigación de riesgo cambiario**

Cobertura cambiaria: Mecanismo para proteger a la empresa de la volatilidad de una moneda por medio de una tarifa previa garantizada acordada previamente.
Las empresas pueden optar a instrumentos que permiten acordar precios con anterioridad usando mercados de futuros. (Agregar certidumbre a esta variable)

El riesgo principal de los contratos con cobertura cambiaria es que haya una sobre estimación del valor del tipo de cambio.

**Futuros financieros**

Mercado en que se negocian contratos en que las partes se comprometen a comprar o vender, en una fecha futura, un determinado activo que pueden ser bienes físicos (comoddities), monedas e instrumentos financieros, a un valor que se fija en el momento.

Cobertura de riesgo de variabilidad de precios: Los inversionistas pueden acordar un precio hoy al que negociarán sus operaciones en el futuro, disminuyendo el riesgo de pérdidas por la baja en los precios en los bienes que forman su cartera.

**Inversionistas de Largo Plazo o Hedgers**

- Son inversionistas que desean protegerse de los riesgos causados por las fluctuaciones en los precios de los mercados físicos.

- Este inversionista trata de asegurarse un precio hoy, disminuyendo las pérdidas y a la vez, disminuyendo las posibles ganancias por aumento en los precios.

**Especuladores**

Asumen el riesgo de fluctuaciones de precios estipulados por la obtención de ganancia de capital, ante las expectativas de incremento en los precios.

### Ejemplo de un Exportador

Un exportador necesita vender las divisas que ha obtenido de sus ingresos operacionales:

- Si vislumbra que el precio de ellas podría bajar con respecto a la moneda local, puede firmar un contrato de futuro de divisas.

- Se compromete a venderlas a un precio determinado, de tal forma que no se pierda en la conversión de divisa a moneda nacional.

- Cuando el precio cae, la caída en valor del instrumento financiero es parcialmente compensada por las ganancias de la posición a futuro. (NO ME QUEDÓ MUY CLARO)

### Tipos de cobertura

La cobertura será distinta si se analiza una empresa importadora o exportadora. Así, la cobertura puede ser de dos tipos:

- **Venta de futuros**: Tiene por objetivo proteger el valor de los instrumentos financieros que posee el inversionista contra posibles caídas en los precios.

- **Compra de futuros**: Tiene por objetivo proteger de posibles aumentos en los precios de los instrumentos financieros que aún no se han adquirido, pero que sí existe la intención de comprar en un tiempo cercano.

Si una empresa es importadora y exportadora, podría optar por los dos tipos de cobertura. Por ejemplo, créditos en moneda extranjera como ingresos en moneda extranjera.
