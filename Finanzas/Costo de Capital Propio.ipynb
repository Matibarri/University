# Introducción

Las empresas tienen varios usos para sus cash-flows no comprometidos:

1. Retener fondos:

    - Invertir en nuevos proyectos.
    - Aumentar reservas de la empresa.
    
2. Pagar a los accionistas:

    - Pagando dividendos.
    - Recomprando acciones.
    - Entregando dividendos en acciones.
    
Se puede hacer una mezcla y tomar ambas decisiones. Así, una proporción se retiene y otra para los accionistas.

Se entenderá por fuentes aportadas por los propietarios a:

1. La reinversión de los fondos generados por la empresa.
2. Los aportes que hacen los dueños (ejemplo: empresas familiares).
3. La venta de acciones en el caso de S.A

**Los accionistas reciben un pago o un beneficio por poseer acciones de la empresa**. Para recibir a éste beneficio, los accionistas deben aportar dinero a la empresa.

## Comparación entre fuentes de financiamiento

Se usan metodologías parecidas en los cálculos del costo por financiamiento. Sin embargo, hay diferencias sustanciales entre financiamiento vía deuda versus el financiamiento vía capital propio.

|         Atributo        |          Deuda         |         Capital Propio        |
|:-----------------------:|:----------------------:|:-----------------------------:|
|       Retribución       | Interés y Amortización | Utilidades netas y dividendos |
|          Plazo          |      Por contrato      |         Indeterminado         |
|    Propiedad activos    |      Sin propiedad     |     Propiedad proporcional    |
| Obligatoriedad de pagos |      Por contrato      |            Acordado           |
|     Proyección pagos    |         Ciertos        |   Inciertos y probabilístico  |
|      Tipo de moneda     |      Por contrato      |             Pesos             |

En general, estas diferencias están asociadas a que la deuda es un contrato con características pactadas y previamente definidas. No así, con el capital propio, ya que esto es incierta y podría ir cambiando año a año. Así, el proceso real del cálculo del costo de financiamiento por Capital Propio es mediante métodos probabilísticos.

## Costo explícito de financiamiento por acciones

A la empresa, le cuesta el dinero proveniente de los accionistas, una tasa $k_p$ que resuelve:

Dinero recibido por la empresa $=$ Suma descontada de beneficios futuros que recibirán los accionistas y que debe pagarles la empresa (dividendos).

se define cada lado de la igualdad como:

Dinero recibido por la empresa $= P_0(1-c)$

Suma descontada de beneficios futuros $= \sum_{n=1}^{N} \frac{BN_n}{(1+k_p)^n}$

donde:

- $P_0 =$ precio de una acción en el momento que propietario compra acción.
- $c =$ costo de corredores y comisiones de bolsa, $0<c1<$ (costo de corretaje)
- $BN_n =$ beneficio neto por acción en $n$. (Beneficios proyectados)
- $k_p =$ costo de capital propio de accionistas (incógnita del modelo).
- $N =$ período máximo de referencia de los accionistas. Para estos modelos se asume que los inversionistas tienen intereses de empresarios, es decir, mantendrán la propiedad de la empresa por muchos años. El número de período que la empresa le debe pagar al accionista es incierto, pero se asume muy grande (podría tender a infinito)

Así, el costo explícito del capital aportado por los propietarios es la tasa $k_p$ que resuelve la siguiente ecuación:

$P_0(1-c)= \sum_{n=1}^{N} \frac{BN_n}{(1+k_p)^n}$


## Aproximación del costo explícito para accionistas de largo plazo (accionistas empresarios)

Tipos de accionistas:

- Accionistas de largo plazo / empresarios.
- Accionistas de corto plazo / especuladores.

Los accionistas de largo plazo mantienen sus acciones por largos períodos, es decir, no se comportan como especuladores y no están dispuestos a vender sus acciones en el corto plazo. Este tipo de accionistas son los que le interesan a las empresas para realizar el cálculo del costo explícito.

### Accionistas de largo plazo

Desde la perspectiva de los beneficios del inversionista, ellos reciben los beneficios por acción (dividendos) y además reciben en el último período un precio por vender sus acciones ($P_N$).

Dinero que recupera el inversionista $  \sum_{n=1}^{N} \frac{BN_n}{(1+k_p)^n} + \frac{P_N}{(1+k_p)^N} $

Considerar:

- $P_N$ no sale desde la empresa, por eso no se considera en el cálculo de los ejemplos anteriores.

- Además, en el largo plazo, $N \rightarrow \infty$. Por lo tanto, independientemente del precio, se tiene que $\frac{P_N}{(1+k_p)^N} \rightarrow 0$

### Metodología aproximada de valoración de acciones

Los modelos de valoración de acciones normalmente asumen que los beneficios ($BN$) se incrementan a una tasa promedio constante $g >0$ por período durante un horizonte de $n$ años y normalmente se considera que $c=0$. Así, considerando $P_N$ como el precio de la acción en el momento de la venta, se tiene que:

$ P_0 = \frac{BN}{(1+k_p)} + \frac{BN(1+g)}{(1+k_p)^2} + \frac{BN(1+g)^2}{(1+k_p)^3} + ... + \frac{BN(1+g)^{n-1}}{(1+k_p)^n} + \frac{P_N}{(1+k_p)^N} $

Luego de hacer algunos cálculos algebraicos, se obtiene lo siguiente:

$ P_0 = BN( \frac{(1+g)}{(1+k_p)} + \frac{(1+g)^2}{(1+k_p)^2} + \frac{(1+g)^3}{(1+k_p)^3} + ... + \frac{(1+g)^{n}}{(1+k_p)^n} ) \frac{1}{(1+g)} + \frac{P_N}{(1+k_p)^N} $

Suponiendo accionistas de largo plazo, es decir, $n \rightarrow \infty$, entonces:

$ P_0 = \frac{BN}{k_p - g}$

con $k_p \neq g$. Donde $g$ se calcula con datos históricos y normalmente es $ \overline{\Delta BN} \pm \sigma$. Con esto, se pueden proyectar 3 escenarios: El primer escenario con el promedio de los $g$ históricos, un escenario pesimista donde $g$ es el promedio de los crecimientos **menos** su desviación estándar y un escenario optimista donde $g$ es el promedio de los crecimientos, pero esta vez **más** la desviación estándar.

**DUDA**: Dicha aproximación se puede considerar una cota inferior o superior?

Lo que se busca es el costo de capital de los propietarios, y ello se obtiene despejando $k_p$ de la igualdad anterior:

$ k_p = \frac{BN}{P_0} + g $

Este modelo es denominado el **Modelo de Gordon y Shapiro**, la cual muestra que el costo de capital propio de una empresa tiene dos componentes: el costo de capital equivalente al que tiene una empresa que no crece ($\frac{BN}{P_0}$) más la tasa de crecimiento de los beneficios. Dicho de otro modo, la expresión $\frac{BN}{P_0}$ se puede definir como el beneficio por cada peso invertido.

- Los propietarios exigen como mínimo lo que la empresa les promete de beneficios hoy día. ($g=0$)

- Los propietarios además exigen lo que se les pueda prometer de crecimiento de esos beneficios, en un larguísimo plazo.

**Referencia**: Gordon, M.J., & Shapiro, E. (1956). Capital equipment analysis: the required rate of profit. Management Science, 3(1), 102-110.

### Ratio Precio-Utilidad o PER

El indicador precio/utilidad (también denominado PER del inglés Price/Earning Ratio) tiene mucha aplicación en el análisis bursátil.
El PER representa el valor actual de $1$ de beneficio neto por acción se deude directamente del modelo anterior si asumimos $g=0$, entonces:

$k_p = \frac{BN}{P_0}$

Así, suponiendo $c=0$ (costo de corretaje es cero), $g=0$ (la empresa no crece) y $n \rightarrow \infty$ (a largo plazo), se define el PER como:

PER $=$ Ratio Precio/Utilidad $= \frac{P_0}{BN} = \frac{1}{k_p}$

Por lo tanto, el PER es un indicador aproximado de rentabilidad de las acciones comunes, pero tomando el inverso de este indicador. En efecto, si se condiera el costo de capital de los accionistas comunes ($k_p$), entonces esa es la rentabilidad que obtendrían por invertir en acciones de la empresa.

**Nota**: El PER aparece en páginas como Emol inversiones.
